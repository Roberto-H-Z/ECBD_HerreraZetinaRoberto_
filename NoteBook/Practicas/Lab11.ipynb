## Etapa 4: Definición de variables independientes (X) y dependiente (y)

X - Variables que utilizara el

y - Variable que queremos predecir

```python
columnas = [
    "Pclass",
    "Sex",
    "Age",
    "SibSp",
    "Parch",
    "Fare",
    "Embarked"
]
```

```python
#Definimos x y
x = dfTitanic[columnas]
y = dfTitanic["Survived"]
```

## Etapa 5: Manejo de datos nulos o faltantes

Para simplificar el proceso, realizaremos la imputación de los datos de la separación TRAIN/TEST

```python
#Revisamos las variables predictoras
X.isnull().sum()
```

```python
X["Age"] = X["Age"].fillna(
    X["Age"].median()
)
```

```python
X["Embarked"] = X["Embarked"].fillna(
    X["Embarked"].mode()[0]
)
```

```python
X.isnull().sum()
```

```python
X.head()
```

## Etapa 6: Transformación de texto a números (Encoding)
Los algoritmos no entienden texto. Se deben convertir las variables categoricas en variables numericas

```python
X = pd.get_dummies(
    X,
    columns=["Sex","Embarked"],
    drop_first=True
)
```

```python
X.head()
```

## Etapa 7: Separación en conjuntos de entrenamiento y prueba

No debemos entrenar y evaluar utilizando los mismos datos.

La racomendación es :

Entrenamiento 80%

Prueba 20%

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.20,
    random_state = 42
)
```

```python
print("Entrenamiento: ", X_train.shape)
```

```python
print("Prueba:", y_test.shape)
```

## Etapa 8: Estandarización de las características numéricas

Las variables numericas tienen valores distintos. El escalado evita que una variable domine más que otra.

```python
columnas_numericas = [
    "Age",
    "Fare",
    "SibSp",
    "Parch"
]
```

```python
X_train[columnas_numericas].describe()
```

```python
scaler = StandardScaler()
X_train[columnas_numericas] = scaler.fit_transform(
    X_train[columnas_numericas]
)
X_test[columnas_numericas] = scaler.transform(
    X_test[columnas_numericas]
)
```

```python
X_train[columnas_numericas].describe()
```

## Etapa 9: Creación y ajuste del algoritmo
Usaremos como ejemplo un arbol de decision para clasiicar registros

```python
modelo = DecisionTreeClassifier(
    max_depth = 4,
    random_state = 42
)
```

```python
modelo.fit(
    X_train,
    y_train
)
```

## Etapa 10: Ejecución de predicciones
Realizamos predicciones para evaluar el modelo

```python
predicciones = modelo.predict(
    X_test
)
```

```python
pd.DataFrame({
    "Real ": y_test,
    "Prediccion ": predicciones
        }).head(30)
```

## Etapa 11: Cálculo de la exactitud (Accuracy)
El accuracy determina el porcentaje de predicciones es correcta realizadas por el modelo

```python
accuracy = accuracy_score(
    y_test,
    predicciones
)
```

```python
print(f"accuracy : {accuracy: .4f}")
```
