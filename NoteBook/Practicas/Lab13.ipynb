```python
#Importar librerias
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import ( accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay )
```

```python
##Cargamos dataset Iris
from sklearn.datasets import load_iris
iris = load_iris(as_frame=True)
```

```python
print(iris.DESCR)
```

```python
dfIris = iris.frame
```

```python
dfIris.head()
```

```python
## Exploracion de conjunto de datos
dfIris.shape
```

```python
dfIris.columns
```

```python
iris.target_names
```

```python
### Renombrar columna "tarjet"

dfIris.rename(columns={'target':'species'}, inplace=True)
```

```python
dfIris["species"] = dfIris["species"].map({
    0:"setosa", 1:"versicolor", 2:"virginica"
    })
```

```python
dfIris.head()
```

```python
dfIris.tail()
```

```python
# variable predictora

X = dfIris.drop(columns = ["species"])
```

```python
print("Variables predictoras")
print(X.columns)
```

```python
# Variable objetivo

y = dfIris["species"]
```

```python
print("Variables objetivo")
print(y.name)
```

## Dividir los datos en prueba y entrenamiento

80% para entrenamiento

20% para prueba

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)
```

```python
print("X_train", X_train.shape)
print("X_test", X_test.shape)
print("y_train", y_train.shape)
print("y_test", y_test.shape)
```

### Escalado de variables

```python
scaler = StandardScaler()
```

```python
# Escalamos las variables

X_train_escalado = scaler.fit_transform(X_train)
X_test_escalado = scaler.transform(X_test)
```

```python
print("Sin escalar")
print(X_train.head())
print("Escalado")
print(X_train_escalado[:5])
```

## Entrenamiendo del modelo knn

En esta prueba utilizaremos k=3

```python
modelo = KNeighborsClassifier(n_neighbors=3)
```

```python
modelo.fit(X_train_escalado, y_train)
```

### Realizamos predicciones con las variables de prueba

```python
prediccion = modelo.predict(X_test_escalado)
```

```python
print(prediccion[:10])
```

```python
# Comparacion entre valores reales y valores predecibles

comparacion = pd.DataFrame({
    "especie real": y_test.values,
    "especie predicha": prediccion
})
```

```python
print(comparacion.head(10))
```

### Evaluacion del modelo

```python
accuracy = accuracy_score(y_test, prediccion)
print(f"accuracy/exactitud: {accuracy: .4f}")
```

```python
## Matriz de confusion

cm = confusion_matrix(y_test, prediccion)
```

```python
cm_df = pd.DataFrame(cm, index=modelo.classes_, columns=modelo.classes_)
```

## Clasification report

proporciona metricas que permiten evaluar el desempeño de cada clase: precision, recall, f1-score, support


```python
print(classification_report(y_test, prediccion))
```
