```python
#Crear un dataframe con 1000000 de registros
cantidad = 1000000
```

```python
#Datos aleatorios
nombres = ["Ana", "Luis", "Martha", "Carolina", "Sofia", "Pedro", "Elena", "Diego"]
ciudades = ["CDMX", "Monterrey", "Guadalajara", "Puebla", "Cancun", "Merida", "Tijuana", "Toluca"]
categoria = ["Premium", "Básico", "Premium", "Estándar", "Básico", "Estándar", "Básico", "Premium"]
```

```python
import random
```

```python
dfMillon = pd.DataFrame({
    "nombre" : [random.choice(nombres) for _ in range(cantidad)],
    "edad" : [random.randint(18, 70) for _ in range(cantidad)],
    "salario" : [round(random.uniform(5000, 50000), 2) for _ in range(cantidad)],
    "activo" : [random.choice([True, False]) for _ in range(cantidad)],
    "categoria" : [random.choice(categoria) for _ in range(cantidad)],
    "ciudad" : [random.choice(ciudades) for _ in range(cantidad)],
    "anio_registro" : [random.randint(2000, 2026) for _ in range(cantidad)],
})
```
