```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("test.csv")
```

```python
df
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
df.columns
```

```python
#Limpieza de columnas en minusculas

df.columns = df.columns.str.strip().str.lower()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
#Eliminar colummna cabin

df.drop(columns=["cabin"], inplace=True)
```

```python
df.columns
```

```python
df["age"] = df["age"].fillna(df["age"].median())
```

```python
df.isnull().sum()
```

```python
#Rellenamos la columna Fare con el valor mas repetido (moda)

df["fare"] = df["fare"].fillna(df["fare"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["age"].min()
```

```python
df["age"].max()
```

```python
df
```

```python
#Graficamos el total de pasajeros por clase
plt.figure(figsize=(10,5))#Tamaño
sns.countplot(df,x="pclass")
plt.title("Total de pasajeros por clase")
plt.xlabel("Clase")
plt.ylabel("Total")
plt.show()
```

```python
#Graficamos las clases por genero

plt.figure(figsize=(8,4))
sns.countplot(df,x="pclass",hue="sex")
plt.title("Total de pasajeros por clase por genero")
plt.xlabel("Clase")
plt.ylabel("Total")
plt.show()
```

```python
#Mostramos la edad por genero y clase

plt.figure(figsize=(6,4))
sns.violinplot(df,x="sex",y="age",hue="pclass")
plt.title("Edad por genero y clase")
plt.xlabel("Clase")
plt.ylabel("Edad")
plt.show()
```

```python
#CONOCIMINETO
#Los pasajeros de primera clase pagaban tarifas mas altas
df.groupby("pclass")["fare"].mean()
```
